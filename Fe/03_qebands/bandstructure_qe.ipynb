# Band Structure - Quantum Espresso

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
x_values = np.array([0.0000, 1.0000, 1.8660, 2.3660, 3.0731, 4.0731, 4.7802, 5.4873, 6.3534, 6.8534])
x_labels = np.array([r'$\Gamma$', 'H', 'P', 'N', r'$\Gamma$', 'H', 'N', r'$\Gamma$', 'P', 'N'])
```

```python
data = np.genfromtxt('bands.out.gnu')
```

```python
# Define font settings for Arial
arial_font = {'family': 'Arial', 'weight': 'normal', 'size': 8}
```

```python
num_breaks = 0
 
if num_breaks > 0:
    x_unique = np.unique(data[:,0])  
    x = data[:num_breaks+len(x_unique),0]
    y = np.transpose(np.reshape(data[:,1], (-1,len(x))))
if num_breaks == 0:
    x = np.unique(data[:,0])
    y = np.transpose(np.reshape(data[:,1], (-1,len(x))))
```

```python
plt.figure(1)
plt.plot(x, y, color = 'b')

# Create the figure with the specified size
plt.figure(1, figsize=(6.30, 3.94))

plt.hlines(12.6279, x[0], x[-1], linestyles='--', color='k')

# Customize ticks with font and size
plt.xticks(x_values, x_labels, fontsize=8, fontname='Arial')
plt.yticks(fontsize=8, fontname='Arial')
plt.ylabel("Energy (eV)")
plt.xlim(x[0], x[-1])

# Customize ticks and grid
plt.tick_params(direction='in')
plt.grid()

# Save the figure
plt.savefig("NiSi_BandStructure_QE.png", dpi=1200, bbox_inches='tight')

# Display the plot
plt.show()
```
