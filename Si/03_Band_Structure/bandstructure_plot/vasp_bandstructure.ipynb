# Band Structure - Si

```python
import py4vasp
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
```

```python
bandcalc = py4vasp.Calculation.from_path('/Users/dhett/DFT/Si/BandStructure/')
```

```python
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
dict = bandcalc.band.to_dict()
```

```python
x = np.array(dict["kpoint_distances"])
y = np.array(dict["bands"])
#y_up = np.array(dict["bands_up"])
#y_down = np.array(dict["bands_down"])
```

```python
given_labels = np.array(dict["kpoint_labels"])
x_test_labels = np.array([''] * np.size(x))
x_values = []
```

```python
for i in range(np.size(given_labels)):
    if x_test_labels[i] != given_labels[i]:
        x_values.append(x[i])
```

```python
#Path: G-X-W-K-G-L-U-W-L-K U-X  
x_labels = np.array([r'$\Gamma$', 'X', 'X', 'W', 'W', 'K', 'K', r'$\Gamma$', r'$\Gamma$', 'L', 'L', 'U', 'U', 'W', 'W', 'L', 'L', 'K|U', 'K|U', 'X'])
```

```python
plt.figure(1, figsize=(6.30, 3.94))

arialfont = {'fontname':'Arial'}

plt.xticks(x_values,x_labels, fontdict=arialfont, fontsize = 8)
plt.yticks(fontsize = 8)

#plt.plot(x, y_up,'purple', linewidth=0.5)
#plt.plot(x, y_down, 'b', linewidth=0.5)
plt.plot(x, y,'purple', linewidth=0.5)
plt.ylabel("E - $E_{F}$ (eV)", fontdict=arialfont, fontsize = 8)
plt.hlines(0, x[0], x[-1], linestyle = '--', color = 'k', linewidth=0.5)
plt.vlines([x_values[13:17]], -1, 1, linestyle = '-', color = 'k', linewidth=0.5)
plt.xlim(x_values[0], x_values[-1])
#plt.ylim(-1,1)
plt.tick_params(direction='in')
plt.grid()
legend_lines = [Line2D([0], [0], color='purple', linestyle='-', linewidth=0.5), Line2D([0], [0], color='b', linestyle='-', linewidth=0.5)]
#plt.legend(legend_lines, ['spin up', 'spin down'], fontsize = 8, bbox_to_anchor=(1,1))

plt.savefig("Si-BandStructure.png", dpi=1200, bbox_inches = 'tight')
plt.show()
```
