```python
import numpy as np
import matplotlib.pyplot as plt

data_file = 'bandstructure'

data = np.loadtxt(data_file)

k_points = data[:, 0] 
energies = data[:, 4:]  

num_bands = energies.shape[1]

fermi_energy = 0.149482

plt.figure(figsize=(8, 6))

for i in range(num_bands):
    energy = energies[:, i]
    plt.plot(k_points, energy, color='green' if np.any(energy < fermi_energy) else 'blue')

plt.xlim(0,1)
plt.xlabel('K')
plt.ylabel('Energy (eV)')
plt.title('Bulk Silicon Band Structure')

plt.axhline(fermi_energy, color='black', linestyle='--', label=f'Fermi Energy = {fermi_energy} eV')

plt.grid(True)

plt.legend()
plt.show()
```
